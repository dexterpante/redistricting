#### Instructions
1.  **Distance to Municipality/Poblacion:**
    *   **Methodology:** `osmnx.geocode_to_gdf()` is used to retrieve the administrative boundary of the municipality. The query string is formed by combining the `municipality_name` with ', Philippines'. The centroid of this retrieved boundary (`gdf_municipality.geometry.union_all().centroid`) is calculated to represent the central point of the municipality. `geopy.distance.geodesic` is then used to calculate the geodesic distance in kilometers between the school's coordinates (`lat`, `lon`) and the municipality's centroid (`municipality_centroid.y`, `municipality_centroid.x`). The result is rounded to two decimal places.
    *   **'N/A' Criteria:** If `osmnx.geocode_to_gdf()` returns an empty GeoDataFrame (meaning the municipality could not be geocoded) or any exception occurs during geocoding or centroid calculation, the `distance_to_municipality_km` is set to `np.nan`.
2.  **Distance to Major Paved Road:**
    *   **Methodology:** `osmnx.graph.graph_from_point()` is employed to retrieve a road network within a `dist_search_buffer` of 10 km (the `radius_km`) around the school's location. A `custom_filter` is applied to specifically target major paved roads: `[highway~"primary|secondary|trunk|motorway"][surface~"paved"]`. If a road network (`G`) is successfully retrieved and contains edges, the graph is converted into GeoDataFrames for nodes and edges using `ox.utils_graph.graph_to_gdfs()`. The geodesic distance from the `school_point` to each `LineString` segment of the identified roads is approximated by calculating the distance to its start and end points using `geopy.distance.geodesic`. The minimum of these distances across all road segments is taken as the approximate distance to the nearest major paved road. The result is rounded to two decimal places.
    *   **'N/A' Criteria:** If `osmnx.graph.graph_from_point()` fails to find a road network within the search buffer, or if the retrieved network `G` contains no edges (`G.edges` is empty), or if `gdf_edges` is empty, or if any exception occurs during the process, the `distance_to_major_paved_road_km` is set to `np.nan`.
3.  **Distance to Minor Paved Road:**
    *   **Methodology:** The methodology for calculating the distance to minor paved roads is identical to that used for major paved roads. The key difference lies in the `custom_filter` applied to `osmnx.graph.graph_from_point()`, which targets minor paved roads: `[highway~"tertiary|residential|unclassified|service|living_street"][surface~"paved"]`.
    *   **'N/A' Criteria:** The criteria for setting `distance_to_minor_paved_road_km` to `np.nan` are the same as for major paved roads: if no road network is found, or if the network contains no edges, or if `gdf_edges` is empty, or if an exception occurs.

## Define Geospatial Feature and Distance Calculation Function

### Subtask:
Develop a comprehensive Python function that takes a school's latitude, longitude, and a radius (10 km) as input. This function will perform the following actions using `osmnx` and `geopandas`:
1.  **Geographical Features:** Query OpenStreetMap for the presence of mountains (e.g., `natural=mountain`), volcanoes (`natural=volcano`), hills (`natural=hill`), forests (`landuse=forest`), rivers (`waterway=river`), and coasts (`natural=coastline`) within the 10 km radius. For each, it will return 'Yes' or 'No' and a confidence level.
2.  **Distance to Municipality/Poblacion:** Identify the administrative boundary or central point of the school's municipality and calculate the geodesic distance (in km) from the school to this point.
3.  **Distance to Major Paved Road:** Query OpenStreetMap for major paved roads (e.g., `highway=primary`, `highway=secondary` with `surface=paved`) within a reasonable search area and calculate the shortest geodesic distance (in km) from the school to the nearest segment.
4.  **Distance to Minor Paved Road:** Similarly, query for minor paved roads (e.g., `highway=tertiary`, `highway=residential` with `surface=paved`) and calculate the shortest geodesic distance (in km) to the nearest segment.
The function will return a dictionary containing all detected features, their 'Yes'/'No' answers, confidence levels, and the calculated distances.

```python
import osmnx as ox
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import numpy as np

def get_geospatial_info(lat, lon, municipality_name, radius_km):
    results = {}
    school_point = Point(lon, lat)

    # 4. For each of the following natural features (mountains, volcanoes, hills, forests, rivers, coastlines):
    feature_tags = {
        'mountain': {'natural': 'mountain'},
        'volcano': {'natural': 'volcano'},
        'hill': {'natural': 'hill'},
        'forest': [{'landuse': 'forest'}, {'natural': 'wood'}],
        'river': {'waterway': 'river'},
        'coastline': {'natural': 'coastline'}
    }

    for feature_name, tags in feature_tags.items():
        # osmnx.features.features_from_point can take a list of tags
        if isinstance(tags, list):
            gdf_features = ox.features.features_from_point(point=(lat, lon), tags=tags, dist=radius_km)
        else:
            gdf_features = ox.features.features_from_point(point=(lat, lon), tags=tags, dist=radius_km)

        if not gdf_features.empty:
            results[f'{feature_name}_present'] = 'Yes'
            results[f'{feature_name}_confidence'] = 'High'
        else:
            results[f'{feature_name}_present'] = 'No'
            results[f'{feature_name}_confidence'] = 'N/A'

    print("Geospatial feature detection implemented.")
    return results
```

```python
from fuzzywuzzy import fuzz
import numpy as np

def standardize_municipality_name(input_name, standardized_df, threshold=80):
    if pd.isna(input_name):
        return np.nan

    # Normalize input name for comparison
    normalized_input_name = str(input_name).strip().lower()

    best_match_name = np.nan
    best_match_score = -1

    for index, row in standardized_df.iterrows():
        standard_name = str(row['standardized_municipality_name']).strip().lower()
        aliases = [str(alias).strip().lower() for alias in row['alias_names']]

        # Check for exact match first
        if normalized_input_name == standard_name or normalized_input_name in aliases:
            return row['standardized_municipality_name']

        # If no exact match, perform fuzzy matching
        # Fuzzy match with standard name
        score_standard = fuzz.ratio(normalized_input_name, standard_name)
        if score_standard > best_match_score and score_standard >= threshold:
            best_match_score = score_standard
            best_match_name = row['standardized_municipality_name']

        # Fuzzy match with aliases
        for alias in aliases:
            score_alias = fuzz.ratio(normalized_input_name, alias)
            if score_alias > best_match_score and score_alias >= threshold:
                best_match_score = score_alias
                best_match_name = row['standardized_municipality_name']

    return best_match_name

print("Defined standardize_municipality_name function with fuzzy matching logic.")
```
